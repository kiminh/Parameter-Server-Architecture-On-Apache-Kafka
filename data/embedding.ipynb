```python
import pandas as pd
```

```python
df = pd.read_csv('./spam.csv',encoding ='latin1')
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
df = df.rename(columns={"v1": "label", "v2": "text"})
```

```python
df.sample(5)
```

```python
import numpy as np
from sklearn.feature_extraction.text import HashingVectorizer

output_dim = 1024
ngram_range = (1,3)

vectorizer = HashingVectorizer(n_features=output_dim, ngram_range=ngram_range, analyzer="char")
```

```python
feature_col = df.text
embedded_feature_col = vectorizer.transform(feature_col).astype(np.float32)

data = pd.SparseDataFrame(embedded_feature_col).to_dense().fillna(0)
```

```python
data['label'] = df['label']

data['label'] = data['label'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
data.sample(5)
```

```python
data.to_csv('./spam_embedded.csv', index=False)
```

```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
input_columns = [str(i) for i in range(1024)]
```

```python
df_train['Score'] = df_train['Score'].astype(str)
df_test['Score'] = df_test['Score'].astype(str)
```

```python
import datawig

#Initialize a SimpleImputer model
imputer = datawig.SimpleImputer(
    input_columns=input_columns, # column(s) containing information about the column we want to impute
    output_column='Score', # the column we'd like to impute values for
)

#Fit an imputer model on the train data
imputer.fit(train_df=df_train)
```

```python
#Impute missing values and return original dataframe with predictions
imputed = imputer.predict(df_test)
```

```python
imputed.head()
```

```python
from sklearn.metrics import classification_report

print(classification_report(imputed.Score, imputed.Score_imputed))
```
